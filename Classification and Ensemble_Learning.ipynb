# DATA

```python
ROOT_DIR="E:\\BongoDev\\Classification_Analysis"
DATA_DIR=os.path.join(ROOT_DIR,"Data")
DATASET_PATH=os.path.join(DATA_DIR,"housing.csv")

dataset=pd.read_csv(DATASET_PATH)
dataset.head()
```

```python
dataset.columns
```

```python
numerical_cols=dataset.select_dtypes(include="number").columns
categorical_col=dataset.select_dtypes(include=['object','str']).columns

print(numerical_cols)
print(categorical_col)
```

```python
# Preprocessing

from sklearn.preprocessing import StandardScaler,LabelEncoder
scaler=StandardScaler()
label_encoder=LabelEncoder()
```

```python
dataset[numerical_cols]=scaler.fit_transform(dataset[numerical_cols])
dataset[numerical_cols]
```

```python
dataset[categorical_col]=dataset[categorical_col].apply(label_encoder.fit_transform)
dataset.head()
```
